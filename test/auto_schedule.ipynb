```python
import schedule
import threading
import time

def run_scheduler():
    def schedule_job():
        print('hello, world!')
    
    schedule.every().seconds.do(schedule_job)
    
    while 1:
        schedule.run_pending()
        time.sleep(0.5)
    
t = threading.Thread(target=run_scheduler)
t.start()

t.join()
```
